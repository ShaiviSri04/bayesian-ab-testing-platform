```python
customers = customers[
    [
    "customer_id",
    "customer_unique_id",
    "customer_city",
    "customer_state"
    ]
    ]
```

```python
payments = payments[
    [
        "order_id",
        "payment_sequential",
        "payment_type",
        "payment_installments",
        "payment_value"
    ]
]
```

```python
reviews = reviews[
    [
        "order_id",
        "review_score"
    ]
    ]
```

```python
items = items[
    [
        "order_id",
        "price",
        "freight_value"
    ]
]
```

```python
orders["order_purchase_timestamp"]=pd.to_datetime(orders["order_purchase_timestamp"])
orders["order_delivered_customer_date"]=pd.to_datetime(orders["order_delivered_customer_date"])
orders["order_estimated_delivery_date"]=pd.to_datetime(orders["order_estimated_delivery_date"])
orders["order_approved_at"]=pd.to_datetime(orders["order_approved_at"])
orders["order_delivered_carrier_date"]=pd.to_datetime(orders["order_delivered_carrier_date"])
```

```python
#delay feature
orders["delivery_delay_days"] = (
    orders["order_delivered_customer_date"]
    - orders["order_estimated_delivery_date"]
).dt.days
```

```python
#aggregate payments
payments_agg = payments.groupby("order_id").agg(
    {
        "payment_value":"sum",
        "payment_installments":"max",
        "payment_sequential":"max"
    }
).reset_index()
```

```python
#aggregate items
items_agg = items.groupby("order_id").agg(
    {
        "price":"sum",
        "freight_value":"sum"
    }
).reset_index()
```

```python
#payment retry flag
payments_agg["payment_retry_flag"]=(
    payments_agg["payment_sequential"]>1)
```

```python
#approval delay
orders["approval_delay_hours"]=(
    orders["order_approved_at"]-orders["order_purchase_timestamp"]
).dt.total_seconds()/3600
```

```python
#merge customers+orders
df = customers.merge(
    orders,
    on="customer_id",
    how="inner"
)
```

```python
#merge payment new features
df = df.merge(
    payments_agg,
    on="order_id",
    how="left"
)
```

```python
#merge reviews
df = df.merge(
    reviews,
    on="order_id",
    how="left"
)
```

```python
df=df.merge(
    items_agg,
    on="order_id",
    how="left"
)
```

```python
df["payment_retry_flag"] = (
    df["payment_retry_flag"]
    .fillna(0)
    .astype(int)
)
```

Behavioural analytics

```python
purchase_frequency = df.groupby(
    "customer_unique_id")["order_id"].count().reset_index()

purchase_frequency.columns=[
    "customer_unique_id",
    "purchase_frequency"
]
```

```python
total_spend=df.groupby(
    "customer_unique_id")["payment_value"].sum().reset_index()

total_spend.columns=[
    "customer_unique_id",
    "total_spend"
]
```

```python
#avg order value = total revenue/orders
aov=df.groupby("customer_unique_id")["payment_value"].mean().reset_index()

aov.columns=[
    "customer_unique_id",
    "average_order_value"
]
```

```python
df = df.merge(
    purchase_frequency,
    on="customer_unique_id",
    how="left"
)
```

```python
df = df.merge(
    total_spend,
    on="customer_unique_id",
    how="left"
)
```

```python
df = df.merge(
    aov,
    on="customer_unique_id",
    how="left"
)
```

```python
df["customer_segment"] = pd.qcut(
    df["total_spend"],
    q=3,
    labels=["Low Value", "Medium Value", "High Value"]
)
```
